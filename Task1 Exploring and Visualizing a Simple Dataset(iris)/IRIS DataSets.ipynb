```python
sns.scatterplot(x="Sepal_Length", y="Sepal_Width", hue="Class", data=datafram)
```

### Scatter Plot
This plot shows the relationship between Sepal Length and Sepal Width.
Each color represents a different species (Setosa, Versicolor, Virginica).
It help to see how different species are separated or clustered based on these features.

```python
import matplotlib.pyplot as plt
sns.histplot(data=datafram, x="Sepal_Length", kde=True)

plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

#### Histogram of Sepal Length
It shows the distribution of Sepal Length values in the dataset.
and it helps understand how frequently different sepal lengths occur and the overall data distribution.

```python
sns.histplot(data=datafram["Sepal_Width"], color="blue", label="Sepal Width", kde=True)
sns.histplot(data=datafram["Petal_Width"], color="red", label="Petal Width", kde=True)

plt.legend()
plt.show()
```

#### Histogram Comparison (Sepal Width & Petal Width)
Compares the distribution of Sepal Width and Petal Width.
And it shows differences between features and helps understand which feature varies more.

```python
sns.boxplot(data=datafram, y="Sepal_Length")

plt.title("Box Plot of Sepal Length")
plt.show()
```

#### Box Plot of Sepal Length:
IT helps detect outliers and spread of the data and it shows minimum, maximum, median, and quartiles of Sepal Length.

```python
sns.boxplot(x="Class", y="Sepal_Length", data=datafram)
plt.title("Sepal Length by Species")
```

#### Box Plot by Species
Shows Sepal Length distribution for each species.
Helps compare how Sepal Length differs between Setosa, Versicolor, and Virginica.
